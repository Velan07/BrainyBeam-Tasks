```python
import pandas as pd
```

```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Sales': [250, 120, 340, 150, 400, 210, 120, 380],
    'Quantity': [10, 5, 8, 3, 15, 7, 6, 9]
}
```

```python
df = pd.DataFrame(data)
```

```python
def standardize_sales(group):
    # Standardize the Sales column (z-score: (x - mean) / std)
    group['Sales_Standardized'] = (group['Sales'] - group['Sales'].mean()) / group['Sales'].std()
    return group
```

```python
result = df.groupby('Region').apply(standardize_sales)
```
