```python
import pandas as pd
```

```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Sales': [250, 120, 340, 150, 400, 210, 120, 380],
    'Quantity': [10, 5, 8, 3, 15, 7, 6, 9]
}
```

```python
df = pd.DataFrame(data)
```

```python
def sales_range(series):
    return series.max() - series.min()
```

```python
grouped = df.groupby('Region').agg({
    'Sales': ['mean', 'sum', sales_range],
    'Quantity': ['sum']
})
```

```python
grouped.columns = ['Average Sales', 'Total Sales', 'Sales Range', 'Total Quantity']
```

```python
print(grouped)
```
